# EDA - Coin Market Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
with open('../coin_data/coins_market_data.json') as f:
    data = json.load(f)
```

## EDA

```python
number_of_dates_btc = len(data['BTC'])      # Start date:   28.04.2013 (2013-04-28)
number_of_dates_eth = len(data['ETH'])      # Start date:   28.04.2013 (2013-04-28)
number_of_dates_xrp = len(data['XRP'])      # Start date:   07.08.2015 (2015-08-07)
number_of_dates_ltc = len(data['LTC'])      # Start date:   04.08.2013 (2013-08-04)

print("Number of dates for BTC: ", number_of_dates_btc)
print("Number of dates for ETH: ", number_of_dates_eth)
print("Number of dates for XRP: ", number_of_dates_xrp)
print("Number of dates for LTC: ", number_of_dates_ltc)
```

```python
max_price_btc = 0
max_price_date_btc = ""
unique_prices_btc = set()

for date, values in data['BTC'].items():
    if values['current_price'] > max_price_btc:
        max_price_btc = values['current_price']
        max_price_date_btc = date
    unique_prices_btc.add(values['current_price'])

number_of_duplicate_prices_btc = number_of_dates_btc - len(unique_prices_btc)

print(f'Max price for BTC at {max_price_date_btc} : {max_price_btc}\n')

print(f'Initial number of prices for BTC: {number_of_dates_btc}')
print(f'Number of unique prices for BTC: {len(unique_prices_btc)}')
print(f'Number of duplicate prices for BTC: {number_of_duplicate_prices_btc}')
```
